## Overall Rating v/s Different Skill Levels

```python
skill_cols = [col for col in fifa.columns if col not in ('name', 'nationality', 'club', 'overall', 'position', 
                                                         'preferred_foot', 'work_rate', 'jersey_number', 
                                                         'international_reputation', 'weak_foot', 'skill_moves', 
                                                         'wage_eur', 'age', 'height_cm', 'weight_kg')]

# jersey_number, international_reputation, weak_foot, skill_moves are categorical
# wage_eur, age, height_cm, weight_kg will need different scales for the axes

plt.figure(figsize=(20, 80))

i = 1
for col in skill_cols:
    plt.subplot(8, 2, i, title=f'overall v/s {col}')
    plt.ylim(0, 100)
    plt.xlim(0, 100)
    sns.scatterplot(data=fifa, y='overall', x=col)
    i = i + 1
```
